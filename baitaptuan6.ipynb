```python
#bai9
a = input().split()
b = input().split()
def chuyen_dict(n):
    d = {}
    for item in n:
        key, value = item.split(":")
        d[key] = int(value)
    return d
a = chuyen_dict(a)
b = chuyen_dict(b)
c ={}
for k, v in a.items():
    if k in b:
        c[k] = v + b[k]
    else:
        c[k] = v 

for k, v in b.items():
    if k not in a:
        c[k] = v

print(dict(sorted(c.items())))



```

```python
#bai9
a = input().split()
b = input().split()
def chuyen_dict(n):
    d = {}
    for item in n:
        key, value = item.split(":")
        d[key] = int(value)
    return d
a = chuyen_dict(a)
b = chuyen_dict(b)

for k, v in b.items():
    if k in a:
        a[k] += v  
    else:
        a[k] = v

print(dict(sorted(a.items())))
```

```python
#bai10 ko phai cap so mà la nhung tuple có tổng bằng k (ko bị giới hạn)
from itertools import combinations
a = list(map(int, input("Nhập danh sách số: ").split()))
k = int(input("Nhập số k: "))

b = []


for r in range(1, len(a)+1):
    for combo in combinations(a, r):
        if sum(combo) == k:
            b.append(list(combo))
c = sorted(tuple(x) for x in b)
print(c)

```

```python
#bai10 cap so bằng b
a = list(map(int, input("Nhập danh sách số: ").split()))
b = int(input())
c = []

for i in range(len(a)):
    for j in range(i+1,len(a)):
        if a[i] + a[j] == b:
            c.append((a[i],a[j]))
print(sorted(c))
```
