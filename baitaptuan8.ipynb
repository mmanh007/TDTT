```python
#bai 16
#cach 1
a = list(map(int, input().split()))
a.sort()
b = sorted(set(a))  
dp = []              
i = 0
while i < len(a):
    cnt = 1
    while i + 1 < len(a) and a[i] == a[i+1]:
        cnt += 1
        i += 1
    dp.append(cnt)
    i += 1
cc = []
ok = True
for cnt, x in zip(dp, b):
    if cnt > x:
        cc.append((cnt - x,x))
    elif cnt < x:
        cc.append((cnt,x))
        ok = False
ha = []
if not cc:
    print("YES")
else:
    print("NO",end="(")
    print("bỏ đi",end=" ")
    for m,n in cc:
        ha.append(f"{m} số {n}")
print(",".join(ha),end=")")


#cach2
from collections import Counter
a = list(map(int, input().split()))
cnt = Counter(a)
kp = []
ok = True
for x, y in cnt.items():
    if y > x :
        kp.append((y-x,x))
    elif x > y:
        kp.append((y,x))
        ok = False
ha = []
if not kp:
    print("YES")
else:
    print("NO",end="(")
    print("bỏ đi",end=" ")
    for m,n in kp:
        ha.append(f"{m} số {n}")
print(",".join(ha),end=")")
```

```python
#bai28
import math
def hopso(n):
    if n < 4:
        return False
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return True
    return False
n = int(input())
a = [j for j in range(n) if hopso(j)]
b = []
m = 0
while m < len(a) - 1:
    c = [a[m]]
    while m < len(a) - 1 and a[m+1] == a[m] + 1:
        m += 1
        c.append(a[m])
    if len(c) > 1:
        b.append(c)
    m += 1
if not b:
    print("ko co ")
else:
    maxlist = b[0]
    for lt in b:
        if len(lt) > len(maxlist):
            maxlist = lt
    print(*maxlist)
```

```python
#bai 29
a = list(map(int, input().split()))
max_len = 1
cur_len = 1
start = 0
best_start = 0
for i in range(1, len(a)):
    if a[i] > a[i-1]:
        cur_len += 1
    else:
        if cur_len > max_len:
            max_len = cur_len
            best_start = start
        cur_len = 1
        start = i
if cur_len > max_len:
    max_len = cur_len
    best_start = start
print(*a[best_start:best_start + max_len])
```

```python
#bai 31
s = list(input().strip())
n = len(s)
if "0" in s:
    i = 0
    while i < n and s[i] == '0':
        i += 1
    if i < n:
        for j in range(n-1, i, -1):
            if s[j] == '0':
                s[i], s[j] = s[j], s[i]
                break
    res = "".join(s).lstrip('0')
    print(res if res else "0")
else:
    for h in range(len(s)):
        mins = h
        for k in range(len(s)-1, h, -1):
            if s[k] < s[mins]:
                mins = k
        if mins != h:
            s[h], s[mins] = s[mins], s[h]
            break
    ha = "".join(s)
    print(ha)
```

```python
#bai31
s = [c for c in input() if c.isdigit()]
n = len(s)
for i in range(n):
    mins = -1
    for j in range(n-1, i, -1):
        if s[j] < s[i]:
            if mins == -1 or s[j] < s[mins]:
                mins = j
    if mins != -1:
        s[i], s[mins] = s[mins], s[i]
        break
res = "".join(s).lstrip('0')
print(res if res else "0")
```
