```python
#W11A4
from collections import Counter as ct
def areAnagrams(path1, path2):
    with open(path1) as f1:
        n1 = f1.read().strip()
    with open(path2) as f2:
        n2 = f2.read().strip()
    return ct(n2) == ct(n1)

areAnagrams("TH13-4-1.txt", "TH13-4-2.txt")


```

```python
#W11A5
def count(path, c):
    with open(path) as f:
        input = f.read().strip().lower()
        dem = 0
        for i in input:
            if i == c:
                dem += 1
    return dem

count("TH13-5.txt", "a")
```

```python
#W11A10
def getMoney(path):
    with open(path) as f:
        money = f.read().strip().split()
        return [int(x) for x in money]
    
getMoney("TH13-10.txt")


def rob(houses):
    if houses == []:
        return 0
    if len(houses) == 1:
        return houses[0]
    
    dp = [0] * len(houses)
    dp[0] = houses[0]
    dp[1] = max(houses[0], houses[1])

    for i in range(2, len(houses)):
        dp[i] = max(dp[i-1], dp[i-2] + houses[i])

    return dp[-1]

rob(getMoney("TH13-10.txt"))
    
```

```python
#W11A12
def maximumProduct(path):
    with open("TH13-12.txt") as f:
        n = int(f.readline().strip())
        num = [int(x) for x in f.read().split()]
    num.sort()
    x1 = num[-1] * num[-2] * num[-3]
    x2 = num[0] * num[1] * num[-1]
    return max(x1, x2)

maximumProduct("TH13-12.txt")
```

```python
#W11A13
def nozero(path):
    with open(path) as f:
        n = int(f.readline().strip())
        num = [int(x) for x in f.read().split()]

    not_zero = [x for x in num if x != 0]
    zero_count = num.count(0)

    return not_zero + [0] * zero_count

nozero("TH13-13.txt")
```

```python
#W11A14
def averageTime(path):
    with open(path) as file:
        lines = file.readlines()[1:] 
        times = [float(line.strip().split(',')[1]) for line in lines]
    
    average = sum(times) / len(times)
    return round(average, 2)

averageTime("TH13-14.csv")
```

```python
#W11A15
def findMovies(path):
    result = []
    with open(path) as file:
        lines = file.readlines()[1:]

        for line in lines:
            parts = line.strip().split(',')
            if len(parts) < 4:
                continue

            movie = parts[1].strip()
            description = parts[2].strip().lower()
            rating = float(parts[3].strip())

            if rating >= 8.0 and "boring" not in description:
                result.append(movie)

    return result

findMovies("TH13-15.csv")

```

```python
#W11A16
import math

def computeAreas(path):
    with open(path) as f:
        lines = [line.strip() for line in f if line.strip()]
    
    for i in range(0, len(lines), 2):
        shape = lines[i].split(':')[1].strip().upper()
        data = lines[i+1]

        if shape == 'SQUARE':
            side = float(data.split(':')[1])
            area = side ** 2
        elif shape == 'RECTANGLE':
            w = float(data.split(':')[1].split()[0])
            h = float(data.split('HEIGHT:')[1])
            area = w * h
        elif shape == 'CIRCLE':
            r = float(data.split(':')[1])
            area = math.pi * r ** 2
        else:
            continue

        print(f"{area:.2f}")

computeAreas("TH13-16.txt")
```

```python
#W11A17
def productExceptSelf(path):
    with open(path) as f:
        n = int(f.readline().strip())
        nums = list(map(int, f.readline().strip().split()))

    result = [1] * n

    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]

    right = 1
    for i in range(n-1, -1, -1):
        result[i] *= right
        right *= nums[i]

    print(*result)

    

productExceptSelf("TH13-17.txt")


```

```python
#W11A18
import math

def estimatedTime(path):
    with open(path) as f:
        lines = [line.strip() for line in f if line.strip()]

    src_x = float(lines[0].split('X:')[1].split()[0])
    src_y = float(lines[0].split('Y:')[1])

    des_x = float(lines[1].split('X:')[1].split()[0])
    des_y = float(lines[1].split('Y:')[1])

    velocity = float(lines[2].split(':')[1].split()[0])

    distance = math.sqrt((des_x - src_x)**2 + (des_y - src_y)**2)

    time = distance / velocity
    return round(time, 2)

estimatedTime("TH13-18.txt")
```

```python
#W11A19
import struct

def readBinaryDoubles(path: str):
    with open(path, 'rb') as f:
        while chunk := f.read(8):  
            number = struct.unpack('d', chunk)[0]  
            print(f"{number:.5f}")  


readBinaryDoubles("TH13-19.bin")

```

```python
#W11A20
def imageSmoother(path):
    with open(path, encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]
        M = int(lines[0])
        a = [list(map(int, line.split())) for line in lines[1:M+1]]

    result = []
    for i in range(M):
        row = []
        for j in range(M):
            total = 0
            count = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    ni, nj = i + dx, j + dy
                    if 0 <= ni < M and 0 <= nj < M:
                        total += a[ni][nj]
                        count += 1
            row.append(total // count)
        result.append(row)

    for row in result:
        print(*row)


imageSmoother("TH13-20.txt")
```
