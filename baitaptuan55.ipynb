```python
#bai3
import math
n = int(input())
def snt(n):
    for i in range(2,math.isqrt(n)+1):
        if n % i == 0:
            return 'Yes'
    else:
        return 'No'
print(snt(n))
```

```python
#bai4
import math

n = int(input())

def shh(n):
    if n <= 1:
        return "No"
    
    tong = 1   # 1 luôn là ước (trừ trường hợp n=1)
    
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            tong += i
            if i != n // i:   # tránh cộng trùng nếu i là căn bậc 2
                tong += n // i
    
    if tong == n:
        return "Yes"
    else:
        return "No"

print(shh(n))

```

```python
def may_tinh(a, b, phep):
    if phep == '+':
        return a + b
    elif phep == '-':
        return a - b
    elif phep == '*':
        return a * b
    elif phep == '/':
        if b == 0:
            return "Lỗi: không thể chia cho 0"
        return a / b
    else:
        return "Phép toán không hợp lệ"

print("=== Máy tính bỏ túi mini ===")
print("Nhập 'q' để thoát chương trình")

while True:
    inp1 = input("Nhập số thứ nhất: ")
    if inp1.lower() == 'q':
        break

    inp2 = input("Nhập số thứ hai: ")
    if inp2.lower() == 'q':
        break

    phep = input("Nhập phép toán (+ - * /): ")
    if phep.lower() == 'q':
        break

    a = float(inp1)
    b = float(inp2)

    ketqua = may_tinh(a, b, phep)
    print("Kết quả:", ketqua)
    print("-----------------------------")
```

```python
#bai8
a = list(map(int,input().split()))
b = list(map(int,input().split()))
def khoangcach_hamming():
    dem = 0
    if len(a) != len(b):
        dem += abs(len(a) - len(b))
    else:
        for x,y in zip(a,b):
            if x != y:
                dem += 1
        return dem
print("Khoang cach hamming là :",khoangcach_hamming())
```

```python
#bai 10
def is_dang_cau(a, b):
    for i in range(len(a)):
        if a.count(a[i]) != b.count(b[i]):
            return False
    return True

print(is_dang_cau("abca", "zbxz"))  
print(is_dang_cau("foo", "bar"))    
print(is_dang_cau("paper", "title")) 
print(is_dang_cau("aba", "ccd"))     
```
