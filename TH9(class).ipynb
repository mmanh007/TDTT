```python
class Cylinder:
    def __init__(self, r, h):
        self.r, self.h = r, h
        self.pi = 3.14
    


    def Sbm(self):
        return 2*self.pi*self.r*(self.r + self.h)
    
    def V(self):
        return self.pi*self.r**2 * self.h
    
r, h = map(float, input().split())
htru = Cylinder(r, h)

print(f"{htru.Sbm():.2f} {htru.V():.2f}")
```

```python
class Date:
    def __init__(self, d, m , y):
        self.d, self.m, self.y = d, m , y

    def nam_nh(self):
        return (self.y % 400 == 0) or (self.y %4 == 0 and self.y % 100 != 0)
    
    def hop_le(self):
        d_in_m = [31, 29 if self.nam_nh() else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        if self.m <1 or self.m > 12:
            return False
        if self.d < 1 or self.d > d_in_m[self.m - 1]:
            return False
        return True
    
    def ngay_tiep(self):
        d_in_m = [31, 29 if self.nam_nh() else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        self.d += 1
        if self.d > d_in_m[self.m - 1]:
            self.d = 1
            self.m+=1
            if self.m > 12:
                self.m = 1
                self.y+=1
        
        return Date(self.d, self.m, self.y)
    

date_str = input().strip()
d, m, y = map(int, date_str.split("/"))
date = Date(d, m , y)
if not date.hop_le():
    print("INVALID")
else:
    next_date = date.ngay_tiep()
    print(f"{next_date.d:02d}/{next_date.m:02d}/{next_date.y:02d}")    
        
```

```python

import math
class Point2():
    def __init__(self, x, y):
        self.x, self.y = x, y

    def gtd(self):
        return self.x == 0 and self.y == 0
    
    def Oy(self):
        return self.x == 0 and self.y != 0
    
    def Ox(self):
        return self.x != 0 and self.y == 0
    
    def dist_O(self):
        return math.sqrt(self.x**2 + self.y**2)
    
    def __str__(self):
        if self.gtd(): 
            pos = "M là gốc tọa độ" 
        elif self.Oy(): 
            pos = "M thuộc trục tung" 
        elif self.Ox(): 
            pos = "M thuộc trục hoành" 
        else: 
            pos = "M không thuộc Ox và Oy" 
        return f"Điểm M({self.x}, {self.y}), {pos}, khoảng cách đến O: {self.dist_O():.2f}"

    

x, y = map(int, input().split())
M = Point2(x, y)
print(M)


```

```python
class ShoppingCart():

    def __init__(self):#1
        self.gio_hang = []

    def them_sp(self, ten, gia):#2
        self.gio_hang.append({"ten": ten, "gia": gia})

    def xoa_sp(self, ten):#3
        for sp in self.gio_hang:
            if sp["ten"] == ten:
                self.gio_hang.remove(sp)
                return True
        return False
    
    def gio_hang_rong(self):#4
        return len(self.gio_hang) == 0
    
    def tong_tien(self):#5
        return sum(sp["gia"] for sp in self.gio_hang)
    
    def hien_thi(self):#6
        if self.gio_hang_rong():
            print("Gio hang rong")
        else:
            for sp in self.gio_hang:
                print(f"{sp['ten']}: {sp['gia']}")

    def xoa_het(self):#7
        self.gio_hang.clear()

# --- Phần menu ---
cart = ShoppingCart()

while True:
    print("\nMenu:")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền giỏ hàng")
    print("6. Hiển thị toàn bộ sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Nhập lựa chọn (2-7): ")

    if choice == "2":
        ten = input("Nhập tên sản phẩm: ")
        gia = float(input("Nhập giá sản phẩm: "))
        cart.them_sp(ten, gia)
    elif choice == "3":
        ten = input("Nhập tên sản phẩm cần xóa: ")
        if cart.xoa_sp(ten):
            print("Đã xóa sản phẩm.")
        else:
            print("Không tìm thấy sản phẩm.")
    elif choice == "4":
        if cart.gio_hang_rong():
            print("Giỏ hàng rỗng.")
        else:
            print("Giỏ hàng không rỗng.")
    elif choice == "5":
        print(f"Tổng tiền giỏ hàng: {cart.tong_tien()}")
    elif choice == "6":
        cart.hien_thi()
    elif choice == "7":
        cart.xoa_het()
        print("Đã xóa toàn bộ giỏ hàng.")
    else:
        print("Lựa chọn không hợp lệ.")

    # Hỏi người dùng có muốn thoát không
    exit_choice = input("Có muốn thoát khỏi chương trình không? (yes/no): ")
    if exit_choice.lower() == "yes":
        print("Chương trình kết thúc.")
        break
```
