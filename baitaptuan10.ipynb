```python
#bai1
a = list(map(int,input().split()))
b = list(map(int,input().split()))
def marge(a,b):
    if not a:
        return b 
    if not b:
        return a
    if a[0] <= b[0]:
        return [a[0]] + marge(a[1:],b)
    else:
        return [b[0]] + marge(a,b[1:])
print(*marge(a,b))
```

```python
#bai4
n = int(input())
def hehe(xau1,xau2):
    xau1 = xau1.upper()
    for i in xau1 :
        if i not in xau2:
            xau1 = xau1.replace(i,"")
    if xau2 == xau1:
        return 'YES'
    else:
        return 'NO'
for i in range(n):
    xaua = input("xâu a:").strip()
    xaub = input("xâu b:").strip()
    print(hehe(xaua,xaub))
```

```python
#bai6
n = int(input())
a = list(map(int, input().split()))
i = n - 2
while i >= 0 and a[i] >= a[i + 1]:
    i -= 1
if i < 0:
    a.reverse()
else:
    j = n - 1
    while a[j] <= a[i]:
        a[i], a[j] = a[j], a[i]
        a[i + 1:] = reversed(a[i + 1:])
print(*a)
```

```python
#bai8:
a = input().upper()
b = [0,0]
def is_path_crossing(moves):
    c = [[0,0]]
    for i in moves:
        if i == 'R':
            b[0] += 1
        if i == 'L':
            b[0] -= 1
        if i == 'U':
            b[0] += 1
        if i == 'D':
            b[0] -= 1
        if b in c:
            return 'True'
        else:
            c.append([b[0],b[1]])
    return 'False'
print(is_path_crossing(a))
```

```python
nums = list(map(int,input().split()))
def manh(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > c:
        b, c = c, b
    if e < b:
        return b
    elif e > c:
        return c
    else:
        return e
print(manh(nums))
```

```python
import bisect
nha = list(map(int,input().split()))
lo = list(map(int,input().split()))
def findRadius(nha, lo):
    nha.sort()
    lo.sort()
    maxa = 0
    for x in nha:
        i = bisect.bisect_left(lo, x)
        trai = abs(x - lo[i-1]) if i > 0 else float('inf')
        phai = abs(lo[i] - x) if i < len(lo) else float('inf')
        maxa = max(maxa, min(trai, phai))
    return maxa
print(findRadius(nha, lo))
```
