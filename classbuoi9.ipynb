```python
#bai1
class Cylinder:
    def __init__(self,bankinh,chieucao):
        self.r = bankinh
        self.h = chieucao
    def dientichbemat(self):
        pi = 3.14
        return self.r * 2 * pi * self.h + 2 * pi * self.r ** 2 
    def thetich(self):
        pi = 3.14
        return (4 * pi * self.r ** 3 ) / 3
bankinh,chieucao = map(float,input().split())
manh = Cylinder(bankinh,chieucao)
print(f"dien tich be mat la : {manh.dientichbemat():.2f} , the tich là : {manh.thetich():.2f}")
```

```python
#bai2:
class Date:
    def __init__(self,day,month,year):
        self.d = day
        self.m = month
        self.y = year
    def namnhuan(self):
        return self.y % 400 == 0 or ( self.y % 4 == 0 and self.y % 100 != 0)
    def thang(self):
        if self.m in [1,3,5,7,8,10,12] :
            return 31
        if self.m in [4,6,9,11]:
            return 30
        if self.m == 2:
            return 29 if self.namnhuan() else 28
    def loi(self):
        if self.m <= 0 or self.m > 12:
            return False
        if self.d > self.thang() or self.d <= 0:
            return False
        if self.y <= 0:
            return False
        return True
    def nextday(self):
        self.d += 1
        if self.d > self.thang():
            self.d = 1 
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        return f"{self.d} : {self.m} : {self.y}"
day,month,year = map(int,input().split())
manh = Date(day,month,year)
if not manh.loi():
    print("Ngay thang nam khong hop le")
else:
    print(manh.nextday())
```

```python
#bai2:
class Date:
    def __init__(self,d,m,y):
        self.d=d; self.m=m; self.y=y
    def namnhuan(self):
        return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
    def thang(self):
        if self.m==2: return 29 if self.namnhuan() else 28
        if self.m in [4,6,9,11]: return 30
        return 31
    def loi(self):
        return self.y>0 and 1<=self.m<=12 and 1<=self.d<=self.thang()
    def nextday(self):
        self.d+=1
        if self.d>self.thang():
            self.d=1; self.m+=1
            if self.m>12:
                self.m=1; self.y+=1
        return f"{self.d}/{self.m}/{self.y}"
d,m,y=map(int,input().split())
x=Date(d,m,y)
print(x.nextday() if x.loi() else "Ngay thang nam khong hop le")
```

```python
# bai3
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def goctoado(self):
        return self.x == 0 and self.y == 0
    def tructung(self):
        return self.x == 0 and self.y != 0
    def truchoanh(self):
        return self.y == 0 and self.x != 0
    def distance(self):
        return (self.x**2 + self.y**2) ** 0.5
x, y = map(float, input().split())
manh = Point(x, y)
if manh.goctoado():
    print("Diem nam tai goc toa do")
elif manh.tructung():
    print("Diem nam tren truc tung")
elif manh.truchoanh():
    print("Diem nam tren truc hoanh")
else:
    print("Diem khong nam tren truc nao")
print(f"Khoang cach den goc toa do: {manh.distance():.2f}")
```

```python
#bai4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def set_numbers(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b == 0:
            return "Không thể chia cho 0"
        return self.a / self.b
    def power(self):
        return self.a ** self.b
    def mod(self):
        if self.b == 0:
            return "Không thể chia lấy dư cho 0"
        return self.a % self.b
a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
calc = Calculator(a, b)
while True:
    print("\n===== MENU =====")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia lấy dư")
    print("7. Nhập 2 số mới")
    choice = input()
    if choice == "1":
        print("Kết quả:", calc.add())
    elif choice == "2":
        print("Kết quả:", calc.subtract())
    elif choice == "3":
        print("Kết quả:", calc.product())
    elif choice == "4":
        print("Kết quả:", calc.divide())
    elif choice == "5":
        print("Kết quả:", calc.power())
    elif choice == "6":
        print("Kết quả:", calc.mod())
    elif choice == "7":
        a = float(input("Nhập số a mới: "))
        b = float(input("Nhập số b mới: "))
        calc.set_numbers(a, b)
        print("Đã cập nhật số mới")
    else:
        print("Lựa chọn không hợp lệ!")
    exit_choice = input("Bạn có muốn thoát? ")
    if exit_choice in ["yes", "YES", "Yes"]:
        print("Chương trình kết thúc.")
        break
```
