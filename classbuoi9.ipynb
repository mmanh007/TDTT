```python
#bai7:
import math
class Fraction():
    def __init__(self,tu,mau):
        self.a = tu 
        self.b = mau
    def checkphanso(self):
        return isinstance(self.a, int) and isinstance(self.b, int) and self.b != 0
    def rutgonphanso(self,m,n):
        x = math.gcd(m, n)
        m //= x
        n //= x
        if n < 0:
            m, n = -m, -n
        return Fraction(m,n)
    def __add__(self,other):
        if not  other.checkphanso() or not self.checkphanso():
            return "phân số  ko hợp lệ"
        m,n = 0,0
        if self.b == other.b:
            m = self.a + other.a
            n = self.b
        else:
            m = self.a*other.b + other.a*self.b
            n = self.b * other.b
        return self.rutgonphanso(m,n)
    def __sub__(self,other):
        if not  other.checkphanso() or not self.checkphanso():
            return "phân số  ko hợp lệ"
        m,n = 0,0
        if self.b == other.b:
            m = self.a - other.a
            n = self.b
        else:
            m = self.a*other.b - other.a*self.b
            n = self.b * other.b
        return self.rutgonphanso(m,n)
    def __mul__(self,other):
        m,n = 0,0
        if not  other.checkphanso() or not self.checkphanso():
            return "phân số  ko hợp lệ"
        else:
            m = self.a * other.a
            n = self.b * other.b
        return self.rutgonphanso(m,n)
    def __truediv__(self,other):
        m,n = 0,0
        if not  other.checkphanso() or not self.checkphanso() or other.a == 0:
            return "phân số  ko hợp lệ"
        else:
            m = self.a * other.b
            n = self.b * other.a
        return self.rutgonphanso(m,n)
    def __str__(self):
        return f"{self.a} / {self.b}"
a,b = map(int,input().split())
s1 = Fraction(a,b)
i,j = map(int,input().split())
s2 = Fraction(i,j)
print("1 : cộng")
print("2 : trừ")
print("3 : nhân")
print("4 : chia")
chon = int(input())
if chon == 1:
    print(s1+s2)
if chon == 2:
    print(s1-s2)
if chon == 3:
    print(s1*s2)
if chon == 4:
    print(s1/s2)
```

```python
#bài 11:
import math
class Angle:
    def __init__(self, A, B, C, D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
    def vector(self, P, Q):
        return [Q[0]-P[0], Q[1]-P[1], Q[2]-P[2]]
    def cross(self, a, b):
        return [
            a[1]*b[2] - a[2]*b[1],
            a[2]*b[0] - a[0]*b[2],
            a[0]*b[1] - a[1]*b[0]
        ]
    def dot(self, a, b):
        return a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
    def length(self, v):
        return math.sqrt(self.dot(v, v))
    def angle_between_planes(self):
        AB = self.vector(self.A, self.B)
        AC = self.vector(self.A, self.C)
        BC = self.vector(self.B, self.C)
        BD = self.vector(self.B, self.D)
        n1 = self.cross(AB, AC)
        n2 = self.cross(BC, BD)
        cos_phi = abs(self.dot(n1, n2)) / (self.length(n1) * self.length(n2))
        phi = math.degrees(math.acos(cos_phi))
        return phi
A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))
goc = Angle(A, B, C, D)
print(f"{goc.angle_between_planes():.2f}")
```

```python
#bai11:
#cach2:
class Angle:
    def __init__(self):
        self.matphang = []

    def themdiem(self, x, y, z):
        self.matphang.append([x, y, z])

    def canh(self,matphang):
        a = matphang
        m = [a[1][0]-a[0][0], a[1][1]-a[0][1], a[1][2]-a[0][2]]
        n = [a[2][0]-a[0][0], a[2][1]-a[0][1], a[2][2]-a[0][2]]
        return m,n
    def phaptuyen(self,m,n):
        vtpt=[
            m[1]*n[2] - m[2]*n[1],
            m[2]*n[0] - m[0]*n[2],
            m[0]*n[1] - m[1]*n[0]
        ]
        return vtpt
    
    def tichvohuong(self, u, v):
        return sum(u[i]*v[i] for i in range(3))

    def dodai(self, v):
        return math.sqrt(self.tichvohuong(v, v))
    def goc(self):
        matphang1 = self.matphang[:3]
        matphang2 = self.matphang[1:]
        AB,AC = self.canh(matphang1)
        BC,BD = self.canh(matphang2)
        n1 = self.phaptuyen(AB,AC)
        n2 = self.phaptuyen(BC,BD)
        x = self.tichvohuong(n1,n2)
        y1 = self.dodai(n1)
        y2 = self.dodai(n2)
        cos_phi = abs(x/(y1*y2))
        return math.degrees(math.acos(cos_phi))
manh = Angle()
k = int(input())
for i in range(k):
    x,y,z = map(float,input().split())
    manh.themdiem(x,y,z)
print(f"{manh.goc():.2f}")

```

```python
#bai 13:
import math
class Triangle():
    def __init__(self,canh1,canh2,canh3):
        self.a = canh1
        self.b = canh2
        self.c = canh3
    def doublegetArea(self):
        a, b, c = self.a, self.b, self.c
        if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return math.sqrt(p * (p - a) * (p - b) * (p - c))
canh1,canh2,canh3 = map(int,input("Nhập ll 3 cạnh của tam giác: ").split())
manh = Triangle(canh1,canh2,canh3)
if manh.doublegetArea() == -1:
    print("invalid")
else:
    print(f"{manh.doublegetArea():.2f}")
        
```
