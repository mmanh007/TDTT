```python
#bai1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
#bai3
arr = list(map(int,input().split()))
def BubbleSort(arr):
    for i in range(len(arr)):
        for i in range(len(arr)-1):
            if arr[i] > arr[i+1]:
                arr[i],arr[i+1] = arr[i+1],arr[i]
    return arr
print(BubbleSort(arr))

def SelectionSort(arr):
    for i in range(len(arr)):
        minarr = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[minarr]:
                minarr = j
        arr[i],arr[minarr] = arr[minarr],arr[i]
    return arr
print(SelectionSort(arr))

```

```python
#bai10
nums = list(map(int,input().split()))
m = int(input())
max = 0
for i in range(len(nums)):
    if nums[i] % m > max:
        max = nums[i] % m
    for j in range(i+1,len(nums)):
        nums[i] += nums[j]
        if nums[i] % m > max:
            max = nums[i] % m
print(max)
```

```python
#bai12
def sosanh(a,b):
    mins = min(len(a),len(b))
    for i in range(mins):
        if a[i] < b[i]:     #so sánh 2 từng chữ cái của 2 từ 
            return True
        if a[i] > b[i]:
            return False
    return len(a) < len(b) #neu cac chu cai giong nhau het thì lấy từ ngắn hơn
c = []
d = [ 'A', 'B', 'C', 'D']
for i in range(4):
    siu = input().lower()
    c.append((siu,d[i])) #gán lần lượt các từ đã nhập cho A B C D
for i in range(4):
    for j in range(i+1,4):
        if sosanh(c[j][0],c[i][0]): #so sanh từ thu i với các từ đằng sau 
            c[i],c[j] = c[j],c[i] #nếu từ nào bé hơn thì đảo từ í lên vị trí thứ i
for k in c: 
    print(k[1],end="") #in ra lần lượt chữ cái ứng với các từ

```

```python
#bai13
n = list(map(int,input().split()))
a = {}
for i in n:
    if i in a:
        a[i] += 1
    else:
        a[i] = 1
allkey = list(a.keys())
allkey.sort()
c = []
for j in range(len(allkey)-1):
    if allkey[j] + 1 == allkey[j+1]:
        tong = a[allkey[j]] + a[allkey[j+1]]
        c.append(tong)
print(max(c) if c else 0)
```
