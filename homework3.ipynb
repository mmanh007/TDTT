```python
import math
while True:
    a = int(input())
    if a <= 0:
        print("vui long nhap lai")
        continue
    else:
        b = []
        for i in range(1,math.isqrt(a)+1):
            if a % i == 0:
                b.append(i)
                if i != a // i:
                    b.append(a//i)
        print("cac uoc nguyen duong cua a :",*sorted(b))
        break
#bai6
```

```python
n = int(input("Nhập số (nhập -1 để kết thúc): "))

max_so = n
min_so = n

while True:
    n = int(input("Nhập số (nhập -1 để kết thúc): "))
    if n == -1:
        break
    if n > max_so:
        max_so = n
    if n < min_so:
        min_so = n

print("Số lớn nhất:", max_so)
print("Số nhỏ nhất:", min_so)
#bai14
```

```python
n = int(input("Nhập n: "))
temp = n
chan = 0
le = 0

while temp > 0:
    digit = temp % 10  
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1
    temp //= 10  

print("Chữ số chẵn:", chan)
print("Chữ số lẻ:", le)
#bai16
```

```python
n = int(input("Nhập n: "))
temp = n

if n < 1:
    print("Không phải dạng 3^k")
else:
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        print(n, "là dạng 3^k")
    else:
        print(n, "không phải dạng 3^k")
#bai18
```

```python
A = int(input("Nhập A: "))

f1, f2 = 1, 1
max_fib = 1

while f2 <= A:
    max_fib = f2
    f1, f2 = f2, f1 + f2

print("Số Fibonacci lớn nhất <= A:", max_fib)
#bai19
```

```python
n = int(input("Nhập n: "))
temp = n
count = 0

while temp != 0:
    digit = temp % 10
    if digit in [2, 3, 5, 7]:  # các số nguyên tố 1 chữ số
        count += 1
    temp //= 10

print("Số chữ số là số nguyên tố:", count)
#bai20
```
